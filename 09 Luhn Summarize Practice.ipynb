## **1-2. Luhn Summarize 구현**

```
import numpy as np

# 토큰화 : 어절 별로 토큰화 방안
tokens = []
for line in docs:
    tokens.append(line.split())

tokens[0:2]
```

```
# 전체 어절 개수
total_count = 0
words = []

for line in tokens:
    for word in line:
        total_count += 1
        words.append(word)

print(total_count, words)
```

```
# 어절별 빈도수 확인
import nltk

freq = nltk.FreqDist(words)

freq
```

```
# 어절별 비율 확인
word_rate = {}

for word in words:
    word_rate[word] =freq[word]/total_count

word_rate
```

```
# 중요단어 결정
important_word = {}

for word in words:
    if word_rate[word] < 0.05 and word_rate[word] > 0.01:
        important_word[word] = word_rate[word]

print("=========== 핵심키워드 ==========")
for w in important_word:
    print(w, ":" ,important_word[w])
```

```
# 문장별 중요도 계산
line_important = []

for line in tokens:
    loc_important = []
    important_count = 0
    for loc, words in enumerate(line):
        for im_word in important_word:
            if im_word in words:
                loc_important.append(loc)
                important_count += 1
    try:
        line_important.append((loc_important[-1] - loc_important[0] + 1)**2/important_count)
    except:
        line_important.append(0)

line_important
```

```
# 중요한 문장 출력
important_line = {}

for num, line in enumerate(docs):
    important_line[line] = line_important[num]

sorted(important_line.items(), reverse=True, key=lambda item:item [1])
```

# **2. Class로 구현**

### **2-2. Class를 이용한 Luhn Summarize 구현**

```
import nltk

class luhn_summarize():
    def __init__(self, input_docs):
        self.input_docs = input_docs
        self.docs_split = []
        self.docs = []
        self.tokens = []
        self.total_count = 0
        self.words = []
        self.freq = {}
        self.word_rate = {}
        self.important_word = {}
        self.line_important = []
        self.important_line = {}

    # split을 이용한 문서 전처리
    def split_docs(self):
        docs_split = self.input_docs.split('.')
        # 엔터 제거
        for line in docs_split:
            self.docs.append(line.replace("\n\n", ""))
        return self.docs

    # 토큰화 : 어절 별로 토큰화 방안
    def tokenize_word(self):
        for line in self.docs:
            self.tokens.append(line.split())
        return self.tokens

    # 전체 어절 개수
    def count_tokens(self):
        for line in self.tokens:
            for word in line:
                self.total_count += 1
                self.words.append(word)
        return self.total_count, self.words

    # 어절별 빈도수 확인
    def check_freq(self):
        self.freq = nltk.FreqDist(self.words)
        return self.freq

    # 어절별 비율 확인
    def check_rate(self):
        for word in self.words:
            self.word_rate[word] = self.freq[word] / self.total_count
        return self.word_rate
    
    # 중요단어 결정
    def check_important(self):
        for word in self.words:
            if self.word_rate[word] < 0.05 and self.word_rate[word] > 0.01:
                self.important_word[word] = self.word_rate[word]

        print("=========== 핵심키워드 ==========")
        for w in self.important_word:
            print(w, ":" , self.important_word[w])

    # 문장별 중요도 계산
    def cal_important(self):
        for line in self.tokens:
            loc_important = [] # 문장내 중요 단어 위치
            important_count = 0 # 문장내 중요 단어 개수 파악
            for loc, words in enumerate(line):
                for im_word in self.important_word:
                    if im_word in words:
                        loc_important.append(loc)
                        important_count += 1
            # 문장 내 중요 단어가 없는 경우 0을 부여하고 지나감
            try:
                self.line_important.append((loc_important[-1] - loc_important[0] + 1)**2 / important_count)
            except:
                self.line_important.append(0)
        return self.line_important

    # 중요한 문장 저장
    def save_important(self):
        for num, line in enumerate(self.docs):
            self.important_line[line] = self.line_important[num]
        return self.important_line

    # 자동 실행
    def run(self):
        self.split_docs()
        self.tokenize_word()
        self.count_tokens()
        self.check_freq()
        self.check_rate()
        self.check_important()
        self.cal_important()
        self.save_important()
        
        # 최종 출력물
        print(sorted(self.important_line.items(), reverse=True, key=lambda item:item [1]))
```
